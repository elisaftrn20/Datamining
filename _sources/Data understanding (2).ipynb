# 4. DETEKSI OUTLIER()

## 4.1 Penjelasan

Deteksi outlier merupakan proses krusial dalam analisis data yang bertujuan untuk mengidentifikasi dan menangani titik data yang secara signifikan berbeda dari pola umum dalam kumpulan data. Metode-metode deteksi outlier meliputi pendekatan statistik seperti Interquartile Range (IQR) dan Z-Score, algoritma machine learning seperti Isolation Forest dan One-Class SVM, serta metode jarak seperti K-Nearest Neighbors (KNN) dan Local Outlier Factor (LOF). Tujuan utama dari deteksi outlier adalah membersihkan data, mendeteksi anomali, dan menghindari bias dalam analisis statistik dan pemodelan. Dengan menggunakan metode deteksi outlier yang sesuai, kita dapat meningkatkan kualitas dan validitas analisis data yang dilakukan.

# 5. NAIVE BAYES CLASSIFIER

## 5.1 Penjelasan

Naive Bayes, atau kadang disebut NaÃ¯ve Bayes Classifier, adalah algoritma machine learning probabilistik yang digunakan dalam berbagai macam tugas klasifikasi. Naive Bayes Classifier adalah sebuah algoritma klasifikasi yang didasarkan pada teorema Bayes dengan asumsi sederhana (naive) bahwa setiap fitur dalam dataset adalah independen satu sama lain

<img src="Bayes_rule.png" width="" align="" />


Dimana:

- \(P(C|X)\) adalah posteriori probabilitas dari hipotesis \(C\) (class \(C\)) setelah diberikan data \(X\).
  
- \(P(X|C)\) adalah likelihood, yaitu probabilitas untuk mendapatkan data \(X\) jika hipotesis \(C\) benar.

- \(P(C)\) adalah prior, yaitu probabilitas awal dari hipotesis \(C\) sebelum data \(X\) diamati.

- \(P(X)\) adalah evidence atau bukti, yaitu probabilitas dari data \(X\) secara keseluruhan.

Dengan menggunakan Teorema Bayes, kita dapat membuat prediksi bahwa data \(X\) adalah anggota dari kelas \(C2\) jika dan hanya jika probabilitas \(P(C2|X)\) lebih tinggi daripada semua \(P(C_k|X)\) untuk semua kelas \(k\).


### 5.1.1 Supervised Learning vs Unsupervised Learning

Dalam supervised learning, model dipelajari dari data yang berisi pasangan input-output yang telah ditentukan sebelumnya. Artinya, setiap contoh dalam data latih memiliki label yang sesuai atau target yang ingin diprediksi. Dalam unsupervised learning, model dipelajari dari data yang tidak memiliki label atau target yang ditentukan sebelumnya. Tujuannya adalah untuk menemukan pola atau struktur dalam data tanpa bantuan label. Dengan kata lain, algoritma unsupervised learning harus mempelajari representasi yang berguna dari data tanpa mengetahui jawaban yang benar. 

### 5.1.2 KLASIFIKASI DAN PREDIKSI

A. Klasifikasi

Memprediksi label kelas (baik diskrit maupun kontinu) melibatkan pembangunan model yang dapat mengklasifikasikan data berdasarkan informasi yang diberikan oleh atribut-atributnya. Proses membangun model melibatkan penggunaan data pelatihan yang sudah dilengkapi dengan nilai label kelas sebagai dasar untuk melatih model. Selama proses pelatihan, model belajar pola atau hubungan antara atribut-atribut input dan nilai label kelas yang sesuai. Setelah model dilatih, ia dapat digunakan untuk mengklasifikasikan data baru dengan memberikan prediksi label kelas berdasarkan atribut-atributnya

B. Prediksi

Prediksi melibatkan penggunaan model yang telah dibangun untuk membuat perkiraan atau estimasi terhadap nilai dari suatu variabel target berdasarkan nilai-nilai atribut yang diberikan. Proses ini seringkali terkait dengan variabel target yang bersifat kontinu, seperti prediksi harga saham, suhu, atau jumlah penjualan.

## 5.2 PERHITUNGAN NAIVE BAYES MENGGUNAKAN CODE

```
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
import csv

# Load the iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Initialize the Naive Bayes classifier
nb_classifier = GaussianNB()

# Split the data into training and test sets (80% training, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Train the classifier
nb_classifier.fit(X_train, y_train)

# Predict the classes for the training set
y_test_pred = nb_classifier.predict(X_test)

# Combine features and target into one list
train_data = np.column_stack((X_train, y_train))

# Write the training data to a CSV file
with open('training_data.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    # Write header
    writer.writerow(iris.feature_names + ['Target'])
    # Write data
    writer.writerows(train_data)

training_accuracy = accuracy_score(y_test, y_test_pred)
print("Training Accuracy:", training_accuracy)

sample = np.array([[7.0, 3.2, 4.7, 1.4]])


prediction = nb_classifier.predict(sample)
print('Prediction:', prediction)
```

Hasil "Training Accuracy: 0.9666666666666667" dari model Naive Bayes yang dilatih dengan data iris menunjukkan bahwa model tersebut memiliki tingkat akurasi sekitar 96.67% dalam mengklasifikasikan data. Prediksi [1] merujuk pada kelas atau label dari data iris. Yaitu merujuk pada jenis "versicolor".
